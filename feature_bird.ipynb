```python
import os
import sys
import cv2
import numpy as np
```

```python
image_dir = 'dataset_edited_2/images'
NUM_CLASSES = 2

train_set = []
with open('dataset_edited_2/train.txt') as f:
    for line in f:
        line = line.rstrip()
        l = line.split()
        train_set.append([l[1], image_dir+'/'+l[0]])

test_set = []
with open('dataset_edited_2/test.txt') as f:
    for line in f:
        line = line.rstrip()
        l = line.split()
        test_set.append([l[1], image_dir+'/'+l[0]])
```

```python
GRAYSCALE = 0
detector = cv2.KAZE_create()

print("train start")
dictionarySize = 2
bowTrainer = cv2.BOWKMeansTrainer(dictionarySize)
sift = cv2.xfeatures2d.SIFT_create()
# sift = cv2.SURF(400)

for i, (classId, data_path) in enumerate(train_set):
    gray = cv2.imread(data_path, GRAYSCALE)
    keypoints, descriptors = detector.detectAndCompute(gray, None)
    descriptors = descriptors.astype(np.float32)
    bowTrainer.add(descriptors)
    sys.stdout.write(".")

codebook = bowTrainer.cluster()
print("train finish")
```

```python
print("test start")
matcher = cv2.BFMatcher()
bowExtractor = cv2.BOWImgDescriptorExtractor(detector, matcher)
bowExtractor.setVocabulary(codebook)

correct_num = 0
all_num = 0

for i, (classId, data_path) in enumerate(test_set):
    gray = cv2.imread(data_path, GRAYSCALE)
    keypoints, descriptors = detector.detectAndCompute(gray, None)
    descriptors = descriptors.astype(np.float32)
    bowDescriptors = bowExtractor.compute(gray, keypoints)
    
    className = {"0": "mukudori", "1": "suzume"}
    
    actual = "???"
    if bowDescriptors[0][0] > bowDescriptors[0][1]:
        actual = className["0"]
    elif bowDescriptors[0][0] < bowDescriptors[0][1]:
        actual = className["1"]
    
    result = ""
    if actual == "???":
        result = " => unknown."
    elif className[classId] == actual:
        result = " => success!!"
        correct_num += 1
    else:
        result = " => fail"
    all_num += 1
    
    print("expected: ", className[classId], ", actual: ", actual, result)
print(correct_num, " / ", all_num, " was correct")
```
